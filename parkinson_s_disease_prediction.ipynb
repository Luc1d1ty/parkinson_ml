# Set Up Features and Target

```python
# Get the features and labels
X = Pkns_data.loc[:,Pkns_data.columns!='status'].values[:,1:] #drops the name column because only values
y = Pkns_data.loc[:,'status'].values
```

# Normalize Data

```python
from sklearn.preprocessing import MinMaxScaler

#Scale the features to between -1 and 1
scaler=MinMaxScaler((-1,1))
X=scaler.fit_transform(X)

#print(f"Normlized Features its max : \n{X.max()} \n\nits min : \n{Pkns_data.min()}" )
print(f"Normlized Features its max : \n{X.max()} \n\nits min : \n{X.min()}" )
```

# Split the data into train and test data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

# MODELLING

# 1. A Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(criterion = 'entropy', max_depth = 20, n_estimators = 10000)
rf.fit(X_train,y_train)
```

```python
y_pred_rf = rf.predict(X_test)
y_pred_rf
```

```python
from sklearn.metrics import accuracy_score

score_rf = round(accuracy_score(y_pred_rf,y_test)*100,2)
print("The accuracy score achieved using the Random Forest model is: "+str(score_rf)+" %")
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, classification_report
from yellowbrick.classifier import ConfusionMatrix
from yellowbrick.classifier import ClassPredictionError
from yellowbrick.classifier import ROCAUC
from yellowbrick.style import set_palette
```

```python
rf_cm = ConfusionMatrix(rf, cmap='GnBu')

rf_cm.fit(X_train, y_train)
rf_cm.score(X_test, y_test)
rf_cm.show()
```

# Classification Report

```python
print(classification_report(y_test, y_pred_rf))
```

```python
visualizer = ROCAUC(rf)

set_palette('bold')

visualizer.fit(X_train, y_train)        # Fit the training data to the visualizer
visualizer.score(X_test, y_test)        # Evaluate the model on the test data
visualizer.show()
```

```python
visualizer = ClassPredictionError(rf)

set_palette('pastel')

visualizer.fit(X_train, y_train)
visualizer.score(X_test, y_test)
visualizer.show()
```

# 2. A Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier

max_accuracy = 0

for x in range(200):
    dt = DecisionTreeClassifier(random_state=x)
    dt.fit(X_train,y_train)
    y_pred_dt = dt.predict(X_test)
    current_accuracy = round(accuracy_score(y_pred_dt,y_test)*100,2)
    if(current_accuracy>max_accuracy):
        max_accuracy = current_accuracy
        best_x = x

#print(max_accuracy)
#print(best_x)

dt = DecisionTreeClassifier(random_state=best_x)
dt.fit(X_train,y_train)
```

```python
y_pred_dt = dt.predict(X_test)
y_pred_dt
```

```python
score_dt = round(accuracy_score(y_pred_dt,y_test)*100,2)
print("The accuracy score achieved using the Decision Tree model is: "+str(score_dt)+" %")
```

# 3. A Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression()
lr_model.fit(X_train,y_train)
lr_model
```

```python
y_pred_lr = lr_model.predict(X_test)
y_pred_lr
```

```python
score_lr = round(accuracy_score(y_pred_lr,y_test)*100,2)
print("The accuracy score achieved using the Logistic Regression model is: "+str(score_lr)+" %")
```

# 4. A SVM model

```python
from sklearn.svm import SVC

svm_model = SVC()
svm_model.fit(X_train,y_train)
```

```python
y_pred_svm = svm_model.predict(X_test)
y_pred_svm
```

```python
score_svm = round(accuracy_score(y_pred_svm,y_test)*100,2)
print("The accuracy score achieved using the SVM model is: "+str(score_svm)+" %")
```

# 5. K Nearest Neighbors Model

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train,y_train)
```

```python
y_pred_knn=knn.predict(X_test)
y_pred_knn
```

```python
score_knn = round(accuracy_score(y_pred_knn,y_test)*100,2)
print("The accuracy score achieved using the KNN model is: "+str(score_knn)+" %")
```

# 6. A Naive Bayes Model

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(X_train,y_train)
```

```python
y_pred_nb = nb.predict(X_test)
y_pred_nb
```

```python
score_nb = round(accuracy_score(y_pred_nb,y_test)*100,2)
print("The accuracy score achieved using the Naive Bayes model is: "+str(score_nb)+" %")
```

# 7. A XGBoost Model

```python
import xgboost as xgb

xgb_model = xgb.XGBClassifier(objective="binary:logistic", random_state=42)
xgb_model.fit(X_train, y_train)
```

```python
y_pred_xgb = xgb_model.predict(X_test)
y_pred_xgb
```

```python
score_xgb = round(accuracy_score(y_pred_xgb,y_test)*100,2)
print("The accuracy score achieved using the XGBoost model is: "+str(score_xgb)+" %")
```

# Confusion Matrix

```python
xgb_model_cm = ConfusionMatrix(xgb_model, cmap='GnBu')

xgb_model_cm.fit(X_train, y_train)
xgb_model_cm.score(X_test, y_test)
xgb_model_cm.show()
```

# Classification Report

```python
print(classification_report(y_test, y_pred_xgb))
```

```python
visualizer = ROCAUC(xgb_model)

set_palette('bold')

visualizer.fit(X_train, y_train)        # Fit the training data to the visualizer
visualizer.score(X_test, y_test)        # Evaluate the model on the test data
visualizer.show()
```

```python
visualizer = ClassPredictionError(xgb_model)

set_palette('pastel')

visualizer.fit(X_train, y_train)
visualizer.score(X_test, y_test)
visualizer.show()
```

# Final Scores

```python
scores = [score_lr,score_nb,score_svm,score_knn,score_dt,score_rf,score_xgb] #,score_nn]
algorithms = ["Logistic Regression","Naive Bayes","Support Vector Machine","K-Nearest Neighbors","Decision Tree","Random Forest","XGBoost"]

for i in range(len(algorithms)):
    print("The accuracy score achieved using "+algorithms[i]+" is: "+str(scores[i])+" %")
```

# A COMPARISON OF THE ACCURACY OF THE SEVEN MODELS¶

```python
sns.set(rc={'figure.figsize':(15,8)})
plt.xlabel("Algorithms")
plt.ylabel("Accuracy score")
sns.barplot(x = algorithms, y = scores)
```

# CONCLUSION

Seven models were assessed for their ability to predict Parkinson's disease. The Random Forrest model and XGBoost displayed the best accuracy score (95%) and is thus the best model for predicting whether a person has Parkinson's disease. The KNN and the Decision Tree models had the same accuracy score (92%).
